```python
def quicksort(lista):
  if len(lista) <= 1:
    return lista
  pivote = lista[0]
  izquierda = []
  derecha = []
  for i in range(1, len(lista)):
    if lista[i] < pivote:
      izquierda.append(lista[i])
    else:
      derecha.append(lista[i])
  return quicksort(izquierda) + [pivote] + quicksort(derecha)

numeros = (23, 45, 16, 37, 3, 99, 22)
listaordenada = quicksort(numeros)
print(listaordenada)
```

```python
def mayor(primero, segundo):
  return primero > segundo | primero < segundo

print(mayor(2, 3))
```
